```python
def cal_eff_purity(file, tree, base_filter, N_gen):
    
    variables = ['Dstarp_isSignal','D0_M','Pi0_Prob']
    recon_df = get_pd(file, tree, base_filter, variables)
    
    N_rec = len(recon_df)
    eff = N_rec/N_gen
    
    N_true = len(recon_df[recon_df['Dstarp_isSignal']==1])
    purity = N_true/N_rec
    
    print(f"eff={eff:.4f}, purity={purity:.4f}")
    
    return (eff, purity)
    
```
