```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ROOT
import ctypes
import math
try:
#     plt.style.use('belle2')
    plt.style.use('belle2_serif')
#     plt.style.use('belle2_modern')
except OSError:
    print("Please install belle2 matplotlib style")   
px = 1/plt.rcParams['figure.dpi']

from main.data_tools.extract_ntuples import get_pd, get_np

from main.data_tools.extract_Nevents import get_Nevents
```

```python
file_name = '/share/storage/jykim/storage_b2/storage/reduced_ntuples/MC15ri/etapip_eteeta/MC15ri_etaetapip_loose_v3_241129_rhopeta_morevar/*ccbar.root'
# base_filter = 'Dp_isSignal==1 && Dp_M>0 && Dp_acos_cosAngleBetweenMomentumAndVertexVectorInXYPlane<0.1 && Dp_CMS_p<2.5'
# base_filter = 'Dsp_Dp_M>1.6 & Dsp_Dp_M<2.1'
base_filter = 'Dsp_Dp_M>0'

signal_variables = ["Dsp_Dp_M","Dsp_M"]
tree = 'etapip_gg_rhopeta'
```

```python
pd_df_rhopeta_gg = get_pd(file=file_name, tree=tree,base_filter=base_filter,variables=signal_variables)
```

```python
fig = plt.figure()
ax = fig.subplots()
xrange = (1.6,2.1,71)

xm = np.linspace(*xrange)

df = pd_df_rhopeta_gg
# df2 = pd_df_test2

N_df = len(df)
# N_df2 = len(df2)

plt.hist(pd_df_rhopeta_gg["Dsp_Dp_M"],bins=xm,histtype='step',label=r'Generic ccbar',density=True);

plt.legend(loc="upper right",fontsize=12.5)
# ax.set_title(r'MC15ri 1/ab' )

n, xe = np.histogram(df["Dsp_Dp_M"], bins=50, range=(xrange[0],xrange[1]))

plt.xlabel("$\mathrm{M(\eta \pi^+}) \; [\mathrm{GeV/c^2}]$")
width = xe[1]-xe[0]
plt.ylabel('Entries'+' /' + '$(' + ' '  + "{0:.3f}".format(width).rstrip('0').rstrip('.') + '\mathrm{GeV/c^2})$');

plt.xlim(xrange[0],xrange[1])
plt.tight_layout()
file1_name="MC15ri_Dsp_Dp_M_loose_v3.png"
plt.savefig(file1_name)
print(len(df))
```
