```python
file_name = '/group/belle/users/jaeyoung/storage/01_recon/kshort/kshortpi0_tight_signalMC_recon_mdst_0_8M.root'
base_filter = 'D0_M>1.67 && D0_M<2.06 && kshort_rank_Dstp==1'
signal_variables = ['D0_M','Dstarp_Q','Dstarp_CMS_p',"Dstarp_isSignal","gamma_p","gamma_CMS_p","Pi0_soft_E","gamma_CMS_cosTheta"]
```

```python
pd_df_test = get_pd(file=file_name, tree='kshort',base_filter=base_filter,variables=signal_variables)
```

```python
fig = plt.figure()
ax = fig.subplots()
xrange = (1.67,2.06,51)

xm = np.linspace(*xrange)

df = pd_df_test
plt.hist(df["D0_M"],bins=xm,histtype='step',label=r'$D^0 \to K_S^0 \pi^0$');

plt.legend(loc="upper left")
#ax.set_title(r'1M signal MC' )

n, xe = np.histogram(df["D0_M"], bins=50, range=(xrange[0],xrange[1]))

plt.xlabel("$\mathrm{M(D^0)} \; [\mathrm{GeV/c^2}]$")
width = xe[1]-xe[0]
plt.ylabel('Entries'+' /' + '$(' + ' '  + "{0:.3f}".format(width).rstrip('0').rstrip('.') + '\mathrm{GeV/c^2})$');

plt.xlim(xrange[0], xrange[1])
```
