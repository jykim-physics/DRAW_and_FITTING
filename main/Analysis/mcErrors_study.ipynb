```python
def mcErrors_truth(tag):
    value_list = list()
    true_value = 1
    false_value = 0
    for index, row in pd_ccbar.iterrows(): 
        if bin(int(row['Dstarp_mcErrors']) & int(tag)) == bin(int(tag)):
            value_list.append(true_value)
        else:
            value_list.append(false_value)
    return value_list
```

```python
def add_to_df(pd_df):

    error_tag = [2048,1024,512,256,128,64,32,16,8,4,2,1]

    for i in error_tag:
        col_name = 'mcErrors_' + str(i)

        pd_df[col_name] = mcErrors_truth(int(i))
        # pd_df.insert(i, column=col_name,value = mcErrors_truth(int(i)) )

    return pd_df
```
