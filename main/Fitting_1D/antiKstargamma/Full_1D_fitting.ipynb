```python
fitting_region = (1.67, 2.06)
x = ROOT.RooRealVar("M(D^{0})", "M(D^{0}) [GeV/c^{2}]", fitting_region[0], fitting_region[1], "")
# x.setBins(100)


# rooDataSet = ROOT_df_start.Book(
#     ROOT.std.move(ROOT.RooDataSetHelper("dataset", "Title of dataset", ROOT.RooArgSet(x))), ("D0_M")
# )
# rooDataSet2 = ROOT_df_start.Book(
#     ROOT.std.move(ROOT.RooDataSetHelper("dataset", "Title of dataset", ROOT.RooArgSet(x, y))), ("D0_M", "Dstarp_Q")
# )


mean = ROOT.RooRealVar("mean", "mean of gaussians", 1.83,1.80, 1.85)
# Novo_peak = ROOT.RooRealVar("Novo_peak", "Novo peak", 1.78,1.75,1.85)    
Novo_width = ROOT.RooRealVar("Novo_width", "Novo peak", 0.05,0.001,0.1)    
Novo_tail = ROOT.RooRealVar("Novo_tail", "Novo peak",  0.5,0,1)    

sigma1 = ROOT.RooRealVar("sigma1", "width of gaussians", 0.03,0,0.1)
alpha = ROOT.RooRealVar("alpha", "width of gaussians", 0.5,0.1,5)
n_CB = ROOT.RooRealVar("n_CB", "width of gaussians", 4,0,50)

sigma3 = ROOT.RooRealVar("sigma3", "width of gaussians", 0.03,0,0.1)


# mu =  ROOT.RooRealVar("mu", "mean of gaussians", 1.81,1.80, 1.85)
# lambda1 = ROOT.RooRealVar("lambda1", "Novo peak", 0.5, 0.1, 10) 
# gamma = ROOT.RooRealVar("gamma", "Novo peak", 5, 0, 20) 
# delta  = ROOT.RooRealVar("delta", "Novo peak", 0.5, 0, 3)

bkg_Vpi0_1 = ROOT.RooGaussian("bkg_Vpi0_1", "Signal component 1", x, mean, sigma1)
bkg_Vpi0_2= ROOT.RooNovosibirsk("bkg_Vpi0_2", "Signal component 3", x, mean, Novo_width, Novo_tail)
# bkg_Vpi0_3 = ROOT.RooJohnson("bkg_Vpi0_3", "Signal component 3", x, mean, lambda1, gamma, delta)
bkg_Vpi0_3 = ROOT.RooCBShape("bkg_Vpi0_3", "Signal component 2", x, mean, sigma3, alpha, n_CB)


bkg_Vpi0_1frac = ROOT.RooRealVar("bkg_Vpi0_1frac", "fraction of component 1 in signal", 0.8, 0.0, 1.0)
bkg_Vpi0__sum1 = ROOT.RooAddPdf("bkg_Vpi0__sum1", "Signal1", [bkg_Vpi0_1, bkg_Vpi0_2], bkg_Vpi0_1frac)

bkg_Vpi0_2frac = ROOT.RooRealVar("bkg_Vpi0_2frac", "fraction of component 1 in signal", 0.2, 0.0, 1.0)
bkg_Vpi0 = ROOT.RooAddPdf("bkg_Vpi0", "Signal1", [bkg_Vpi0_3, bkg_Vpi0__sum1], bkg_Vpi0_2frac)

           
    
bkg_remain_mean = ROOT.RooRealVar("bkg_remain_mean", "mean of gaussians", 1.68,1.65, 1.72)
bkg_remain_sigma1 = ROOT.RooRealVar("bkg_remain_sigma1", "width of gaussians", 0.01,0,0.1)

bkg_remain_tau = ROOT.RooRealVar("bkg_remain_tau", "tau",-2, -10,0)
    
bkg_remain_1 = ROOT.RooGaussian("bkg_remain_1", "Signal component 1", x, bkg_remain_mean, bkg_remain_sigma1)
bkg_remain_2 = ROOT.RooExponential("bkg_remain_2", "Signal component 1", x, bkg_remain_tau)    


bkg_remain_1frac = ROOT.RooRealVar("bkg_remain_1frac", "fraction of component 1 in signal", 0.2, 0.0, 1.0)
bkg_remain = ROOT.RooAddPdf("bkg_remain", "Signal1", [bkg_remain_1, bkg_remain_2], bkg_remain_1frac)

                              
sig_mean = ROOT.RooRealVar("sig_mean", "mean of gaussians", 1.87,1.85, 1.88)
# sig_mean2 = ROOT.RooRealVar("sig_mean2", "mean of gaussians", 1.865,1.85, 1.88)

sig_sigma1 = ROOT.RooRealVar("sig_sigma1", "width of gaussians", 0.01,0,0.05)
sig_sigma2 = ROOT.RooRealVar("sig_sigma2", "width of gaussians", 0.02,0,0.05)
# sig_sigma3 = ROOT.RooRealVar("sig_sigma3", "width of gaussians", 0.01,0,0.05)

sig_alpha = ROOT.RooRealVar("sig_alpha", "width of gaussians", 2 ,0,10)
sig_n_CB = ROOT.RooRealVar("sig_n_CB", "width of gaussians", 3,0,10)
# sig_alpha2 = ROOT.RooRealVar("sig_alpha2", "width of gaussians", 0.5,0.1,3)
# sig_n_CB2 = ROOT.RooRealVar("sig_n_CB2", "width of gaussians", 5,0,30)
                              
sig1 = ROOT.RooCBShape("sig1", "Signal component 2", x, sig_mean, sig_sigma1, sig_alpha, sig_n_CB)
sig2 = ROOT.RooGaussian("sig2", "Signal component 1", x, sig_mean, sig_sigma2)
# sig3 = ROOT.RooGaussian("sig3", "Signal component 1", x, sig_mean, sig_sigma3)

# sig2 = ROOT.RooCBShape("sig2", "Signal component 2", x, sig_mean, sig_sigma2, sig_alpha2, sig_n_CB2)
sig1frac = ROOT.RooRealVar("sig1frac", "fraction of component 1 in signal", 0.5, 0.0, 1.0)
# signal_sum1 = ROOT.RooAddPdf("signal_sum1", "Signal1", [sig1, sig2], sig1frac)
signal_model = ROOT.RooAddPdf("signal_model", "Signal1", [sig1, sig2], sig1frac)


# sig2frac = ROOT.RooRealVar("sig2frac", "fraction of component 1 in signal", 0.2, 0.0, 1.0)
# signal_model = ROOT.RooAddPdf("signal_model", "Signal1", [sig3, signal_sum1], sig2frac)

bkg_Vpi_frac_remain = ROOT.RooRealVar("bkg_Vpi_frac_remain", "fraction of component 1 in signal", 0.8, 0.0, 1.0)
bkg_model = ROOT.RooAddPdf("bkg_model", "Signal1", [bkg_Vpi0, bkg_remain], bkg_Vpi_frac_remain)

# sig_bkg_frac = ROOT.RooRealVar("sig_bkg_frac", "fraction of component 1 in signal", 0.01, 0.0, 1.0)
# sig = ROOT.RooAddPdf("sig", "Signal1", [signal_model, bkg_model], sig_bkg_frac)

nsig = ROOT.RooRealVar("nsig","# signal events",5000,0,len(pd_df_Vgamma_ccbar))
nbkg = ROOT.RooRealVar("nbkg","# bkg events",40000,0, len(pd_df_Vgamma_ccbar))
# nbkg2 = ROOT.RooRealVar("nbkg2","# bkg events",300,0, len(pd_Vgamma_ccbar))


# extended_model = ROOT.RooAddPdf("extended_model", "x_model", ROOT.RooArgSet(sig_model,bkg1_xy_model,bkg2_xy_model,bkg3_xy_model), ROOT.RooArgSet(nsig, nbkg1, nbkg2, nbkg3))
extended_model = ROOT.RooAddPdf("extended_model", "x_model", ROOT.RooArgSet(signal_model,bkg_model), ROOT.RooArgSet(nsig, nbkg))



data = ROOT.RooDataSet.from_numpy({"M(D^{0})": pd_Vgamma_ccbar['D0_M']}, [x])

# data = rooDataSet

canv = ROOT.TCanvas("Canvas", "Canvas", 700, 640)

xlow = ctypes.c_double()
ylow = ctypes.c_double()
xup = ctypes.c_double()
yup = ctypes.c_double()

canv.GetPad(0).GetPadPar(xlow, ylow, xup, yup)
canv.Divide(1,2)

xlow = xlow.value
ylow = ylow.value
xup = xup.value
yup = yup.value

upPad = canv.GetPad(1)
upPad.SetPad(xlow, ylow+0.25*(yup-ylow),xup,yup)

dwPad = canv.GetPad(2)
dwPad.SetPad(xlow, ylow,xup,ylow+0.25*(yup-ylow))

# r = sig.fitTo(data,NumCPU=12, Range=(1.67, 2.06))
r = extended_model.fitTo(data,NumCPU=12, Extended=ROOT.kTRUE)
# r = sig.fitTo(data,NumCPU=12, Range=fitting_region)

# r.Print()
canv.cd(1) 
# frame = x.frame(Title="D^{0} #rightarrow #bar{K}^{*0} #pi^{0}")
frame = x.frame(Title=" ")

data.plotOn(frame,Name = "data1", XErrorSize=0)

extended_model.plotOn(frame, Name="gauss1", Components=signal_model, LineStyle=ROOT.kDashed, LineColor='r')
extended_model.plotOn(frame, Name="CB1", Components=bkg_remain, LineStyle=ROOT.kDashed, LineColor=ROOT.kGreen)
extended_model.plotOn(frame, Name="CB1", Components=bkg_Vpi0, LineStyle=ROOT.kDashed, LineColor=593)
# extended_model.plotOn(frame, Name="CB1", Components=bkg_Vpi0_2, LineStyle=ROOT.kDashed, LineColor=ROOT.kPink)

# sig.plotOn(frame, Name="Novo", Components=sig3, LineStyle=ROOT.kDashed, LineColor='r')
# 
extended_model.plotOn(frame, Name="fitting")


# sig.paramOn(frame)
frame.GetXaxis().SetTitleSize(0.047)
frame.GetXaxis().CenterTitle(True)
frame.GetYaxis().SetTitleSize(0.04)
frame.GetYaxis().SetTitleOffset(1.2)
frame.Draw("PE")


leg1 = ROOT.TLegend(0.75, 0.75, 0.9, 0.9)
leg1.SetFillColor(ROOT.kWhite)
# leg1.SetHeader("The Legend title","C")
leg1.AddEntry("data1", "MC", "PE")
leg1.AddEntry("fitting", "fit", "l")
leg1.Draw()

hpull = frame.pullHist()
hpull.SetFillStyle(1001)
hpull.SetFillColor(1);
for i in range(0,hpull.GetN()):#(int i=0;i<hpull.GetN();++i): 
    hpull.SetPointError(i,0.0,0.0,0.0,0.0)
# pullplot = x.frame(Title(" "))
pullplot = x.frame()
pullplot.SetTitle("")
pullplot.addPlotable(hpull,"BE")
# pullplot.addPlotable(hpull,"PE")

pullplot.SetYTitle("Pull")
pullplot.GetXaxis().SetTitleSize(0)
pullplot.GetYaxis().SetTitleSize(0.17)
pullplot.GetYaxis().CenterTitle(True)
pullplot.GetYaxis().SetTitleOffset(0.2)
pullplot.SetMinimum(-4.)
pullplot.SetMaximum(4.)
pullplot.GetXaxis().SetLabelSize(0.1)
pullplot.GetYaxis().SetLabelSize(0.09)
canv.cd(2)
pullplot.Draw()

xmin1 = ctypes.c_double(fitting_region[0])
xmax1 = ctypes.c_double(fitting_region[1])
# xmin1 = 0.1
# xmax1 = 0.18
line = ROOT.TLine(xmin1,0.0,xmax1,0.0)
line1 = ROOT.TLine(xmin1,3.0,xmax1,3.0)
line2 = ROOT.TLine(xmin1,-3.0,xmax1,-3.0)

line.SetLineColor(ROOT.kRed)
line.SetLineWidth(3)
line1.SetLineColor(ROOT.kRed)
line2.SetLineColor(ROOT.kRed)
line1.SetLineStyle(2)
line2.SetLineStyle(2)
line.Draw("SAME")
line1.Draw("SAME")
line2.Draw("SAME")

canv.Update()

canv.Draw()
canv.SaveAs("test.png")
```
